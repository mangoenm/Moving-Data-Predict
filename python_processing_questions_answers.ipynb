## Moving Big Data Predict: Python Processing MCQ

Use the CSV output file generated by your pipeline to answer the below questions

```python
import numpy as np
import pandas as pd
```

```python
# Define column names
# added columns because csv does not have headers ???
column_names = ['stock_date', 'open_value', 'high_value', 'low_value', 'close_value','volume_traded','daily_percent_change', 'value_change', 'company_name']

# Read the CSV file and store its contents in a DataFrame
df = pd.read_csv('C:/Users/Naomi/Desktop/data/Output/historical_stock_data.csv', names=column_names)

# Display the first few rows of the DataFrame
print(df.head(5))
```

#### 1. What is the most common data type of the columns in your processed data?

```python
# Get data types of each column
data_types = df.dtypes

# Count occurrences of each data type
data_type_counts = data_types.value_counts()

# Get the most common data type
most_common_data_type = data_type_counts.idxmax()

print("Most common data type:", most_common_data_type)
```

#### 2. How many rows does the output dataframe have?

```python
data_type_counts = pd.DataFrame(df)

# Get the number of rows in the DataFrame
num_rows = len(data_type_counts)

print("Number of rows:", num_rows)
```

#### 3. What is the sum of the value_change column?

```python
# Sum of value_change
column_sum = df['value_change'].sum()

print("Sum of value_change:", column_sum)
```

#### 4. What is the mean of the value_change column?

```python
# Mean of value_change
column_mean = df['value_change'].mean()

print("Mean of value_change:", column_mean)
```

#### 5. What is the mean open_value? 

```python
# Mean of open_value
column_mean = df['open_value'].mean()

print("Mean of open_value:", column_mean)
```

#### 6. What is the mean close_value?

```python
# Mean of close_value
column_mean = df['close_value'].mean()

print("Mean of close_value:", column_mean)
```

#### 7. What are the minimum and maximum values in the stock_date column?

```python
# Minimum and maximum values in 'stock_date' column
min_stock_date = df['stock_date'].min()
max_stock_date = df['stock_date'].max()

print("Minimum stock_date:", min_stock_date)
print("Maximum stock_date:", max_stock_date)
```

#### 8. How many distinct values are in the company_name column?

```python
# Count distinct values in 'column1'
distinct_count_company_name = df['company_name'].nunique()

print("Distinct count in company_name:", distinct_count_company_name)
```
