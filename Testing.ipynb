```python
import os

def get_company_name(file_path):
    # Get the filename from the file path
    file_name = os.path.basename(file_path)
    
    # Split the filename by a delimiter if necessary
    # Example: split by underscore ('_') if the company name is separated by underscores
    parts = file_name.split('_')
    
    # Extract the company name from the filename parts
    # Example: if the company name is the first part of the filename
    company_name = parts[0]
    
    return company_name

# Example usage:
file_path = '/path/to/your/csv_files/company_name_data.csv'
company_name = get_company_name(file_path)
print("Company Name:", company_name)
```
