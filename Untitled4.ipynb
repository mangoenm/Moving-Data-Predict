```python
#%pip install psycopg2

import psycopg2

def upload_data_to_postgres(data, table_name, dbname, user, password, host, port):
#def upload_data_to_postgres(data, table_name, dbname, user, password, host='your_host', port='5432'):
    """
    Uploads data to a PostgreSQL database.

    Args:
        data (list of tuples): The data to be uploaded, where each tuple represents a row.
        table_name (str): The name of the table to upload the data into.
        dbname (str): The name of the PostgreSQL database.
        user (str): The username for database authentication.
        password (str): The password for database authentication.
        host (str, optional): The hostname of the PostgreSQL server. Defaults to 'your_host'.
        port (str or int, optional): The port number of the PostgreSQL server. Defaults to '5432'.

    Returns:
        None
    """
    try:
        # Establish a connection to the PostgreSQL database
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)
        
        # Create a cursor object using the connection
        cursor = conn.cursor()

        # Prepare SQL statement for inserting data
        insert_query = "INSERT INTO {} VALUES %s".format(table_name)

        # Execute the insert operation
        psycopg2.extras.execute_values(cursor, insert_query, data)

        # Commit the transaction
        conn.commit()

        print("Data uploaded successfully!")

    except (Exception, psycopg2.DatabaseError) as error:
        print("Error:", error)

    finally:
        # Close the cursor and connection
        if cursor:
            cursor.close()
        if conn:
            conn.close()


```

```python
import psycopg2

def upload_to_postgres(**kwargs):
    """
    Uploads data from a CSV file to a PostgreSQL table.

    Args:
        **kwargs: Arbitrary keyword arguments:
            csv_file (str): Path to the CSV file.
            table_name (str): The name of the table to upload the data into.
            dbname (str): The name of the PostgreSQL database.
            user (str): The username for database authentication.
            password (str): The password for database authentication.
            host (str, optional): The hostname of the PostgreSQL server. Defaults to 'localhost'.
            port (str or int, optional): The port number of the PostgreSQL server. Defaults to '5432'.

    Returns:
        None
    """
    try:
        # Establish a connection to the PostgreSQL database
        conn = psycopg2.connect(
            dbname=kwargs.get('dbname'),
            user=kwargs.get('user'),
            password=kwargs.get('password'),
            host=kwargs.get('host', 'localhost'),
            port=kwargs.get('port', '5432')
        )
        
        # Create a cursor object using the connection
        cursor = conn.cursor()

        # Generate SQL COPY command to import data from CSV file
        copy_sql = """
                    COPY {} FROM %s WITH (
                        FORMAT CSV,
                        HEADER true,
                        DELIMITER ','
                    )
                    """.format(kwargs.get('table_name'))

        # Execute the SQL COPY command
        with open(kwargs.get('csv_file'), 'r') as f:
            cursor.copy_expert(sql=copy_sql, file=f)

        # Commit the transaction
        conn.commit()

        print("Data uploaded successfully!")

    except (Exception, psycopg2.DatabaseError) as error:
        print("Error:", error)

    finally:
        # Close the cursor and connection
        if cursor:
            cursor.close()
        if conn:
            conn.close()

# Example usage:
upload_to_postgres(
    csv_file='data.csv',
    table_name='example_table',
    dbname='your_database',
    user='your_username',
    password='your_password',
    host='localhost',
    port='5432'
)
```

```python
import psycopg2

def execute_sql_script(file_path, dbname, user, password, host, port = '5432'):
    """
    Executes an SQL script file on a PostgreSQL database.

    Args:
        file_path (str): The path to the SQL script file.
        dbname (str): The name of the PostgreSQL database.
        user (str): The username for database authentication.
        password (str): The password for database authentication.
        host (str, optional): The hostname of the PostgreSQL server. Defaults to 'localhost'.
        port (str or int, optional): The port number of the PostgreSQL server. Defaults to '5432'.

    Returns:
        None
    """
    try:
        # Establish a connection to the PostgreSQL database
        conn = psycopg2.connect(
            dbname=dbname,
            user=user,
            password=password,
            host=host,
            port=port
        )

        # Create a cursor object using the connection
        cursor = conn.cursor()

        # Read SQL script from file
        with open(file_path, 'r') as file:
            sql_script = file.read()
            print(sql_script)

        # Execute SQL script
        cursor.execute(sql_script)

        # Commit the transaction
        conn.commit()

        print("SQL script executed successfully!")

    except (Exception, psycopg2.DatabaseError) as error:
        print("Error:", error)

    finally:
        # Close the cursor and connection
        if cursor:
            cursor.close()
        if conn:
            conn.close()

    return "CSV Uploaded to postgres database"
```
